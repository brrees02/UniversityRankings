```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
top_fifteen_universities_df = pd.read_csv("World University Rankings 2023 3.csv")
top_universities_df = top_fifteen_universities_df.reset_index()
top_universities_df.columns = ['Null', 'Test', 'University Rank', 'Name of University' 'Industry Income Score', 'Tuition Rates 2023']

plt.figure(figsize=(10, 6))
bar_chart = sns.barplot(x='Industry Income Score', y='Tuition Rates 2023', data= top_fifteen_universities_df, palette="flare")
bar_chart.set_title('Top 13 University Tuition and Industry Outcomes')
bar_chart.set_xlabel('Industry Income Score')
bar_chart.set_ylabel('Tuition Rates of 2023 in Thousands')
plt.tight_layout()
```

```python
plt.scatter(top_fifteen_universities_df['Industry Income Score'], top_fifteen_universities_df['Tuition Rates 2023'])
plt.title('Tuition Rates 2023 vs. Industry Income Score')
plt.xlabel('Industry Income Score')
plt.ylabel('Tuition Rates 2023')
plt.show()
```
